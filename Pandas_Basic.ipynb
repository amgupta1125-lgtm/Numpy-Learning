```python
import pandas as pd

# Most Imp for Pandas DataFrame - Belove code works as Dictionary :)

data = {
    "Name": ["Abhimanyu", "Satyam", "Divyesh"],
    "Age": [20, 21, 18]
}

df = pd.DataFrame(data)
print(df)
```

```python
# This Function in Python

def greet(name):                          #def ka matlab → function banana
    print("Hello", name)
greet("Abhimanyu")                        #greet → function ka naam


 # For Pandas Example

def add_20(x):                   #def → function banane ke liye # add_20 → function ka naam
    return x + 20                 # x → ek value hai (Age ka number)  # return x + 20 → us value me 20 add karke wapas dega

df["New_Age"] = df["Age"].apply(add_20)  # df → DataFrame (table) & "Age" → Age column & .apply() → work on every value
print(df)
```

```python
# Methode   --> list methode
nums = [1, 2, 3]
nums.append(4)  # for adding
nums.remove(2)  # for REmoving
print(nums)
```
