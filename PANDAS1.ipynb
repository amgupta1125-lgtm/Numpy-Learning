```python
import pandas as pd

data = {
        "name": ['Rahul','Shyam','Abhimanyu','Satyam','Divyesh'],
        "age" : [20,25,36,21,18] ,
        "gender" : ['male','male','male','male','male'],
        "city" : ['Rajkot','pak','Ukren','Uk','Rajkot'],
        "Salary": [20000,25000,36000,21000,18000],
        "Score": [20000,25000,36000,21000,18000],
        "Saving": [10000,15000,26000,11000,8000]
        }
df = pd.DataFrame(data)
print(df)

df.drop(columns =["Salary","city","age","city"], inplace=True)
print(df)
```

```python
""""
NaN = Not a Number
None = (for Object data type)

isnull()

True-- NaN is missing
False-- Value is Present
"""
import pandas as pd

data = {
        "name": ['Rahul',None,'Abhimanyu','Satyam','Divyesh'],
        "age" : [20,None,36,21,18] ,
        "gender" : ['male',None,'male','male','male'],
        "city" : ['Rajkot',None,'Ukren','Uk','Rajkot'],
        "Salary": [20000,None,36000,21000,18000],
        "Score": [20000,None,36000,21000,18000],
        "Saving": [10000,None,26000,11000,8000]
        }

df=pd.DataFrame(data)

print(df.isnull())         #Null value showing True

print(df.isnull().sum())   # sum of null  value in each column

#df.fillna(0, inplace=True)   #(Null = 0)
#print(df)

df['age'].fillna(df['age'].mean(),inplace=True )
print(df)
```

```python
#dropna()
df.dropna(inplace=True)
print(df)
```

```python
#fillna()

df.fillna(0, inplace=True)
print(df)
```
