```python
# Array = collection of type of Data

import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
```

```python
# Basic
print(np.zeros(5))
print(np.ones((2,3)))
print(np.arange(1, 10, 2))
print(np.linspace(1, 10, 5))
```

```python
# Size , Shape , Dimension

c = np.array([[[1,2],[3,4]],
              [[5,6],[7,8]]])
```

```python
print(c)
```

```python
print(c.size)
```

```python
print(c.shape)
```

```python
print(c.ndim)
```

```python
 # Indexing & Slicing
a = np.array([10,20,30,40,50])

print(a[0])
print(a[1:4])
```

```python
 # 2d Slicing
b = np.array([[1,2,3],
              [4,5,6]])

print(b[0,2])
```

```python
# Mathematics in Numpy
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a * b)
print(a ** 2)
print(b**2)
```

```python
# Statics in Numpy

a = np.array([10,20,30,40])

print(np.mean(a))
print(np.sum(a))
print(np.max(a))
print(np.min(a))
print(np.std(a))
```
