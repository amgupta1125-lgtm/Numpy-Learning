```python
# Array = collection of type of Data

import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
```

```python
# Basic
print(np.zeros(5))
print(np.ones((2,3)))
print(np.arange(1, 10, 2))
print(np.linspace(1, 10, 5))
```

```python
# Size , Shape , Dimension

c = np.array([[[1,2],[3,4]],
              [[5,6],[7,8]]])
```

```python
print(c)
```

```python
print(c.size)
```

```python
print(c.shape)
```

```python
print(c.ndim)
```

```python
 # Indexing & Slicing
a = np.array([10,20,30,40,50])

print(a[0])
print(a[1:4])
```

```python
 # 2d Slicing
b = np.array([[1,2,3],
              [4,5,6]])

print(b[0,2])
```

```python
# Mathematics in Numpy
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a * b)
print(a ** 2)
print(b**2)
```

```python
# Statics in Numpy

a = np.array([10,20,30,40])

print(np.mean(a))
print(np.sum(a))
print(np.max(a))
print(np.min(a))
print(np.std(a))
```

```python
# Reshape
a = np.array([1,2,3,4,5,6])
b = a.reshape(2,3)
print(b)
```

```python
# BroadCasting (imp**)
a = np.array([1,2,3])
print(a + 10)
```

```python
# Boilean Indexing

a = np.array([10,20,30,40])

print(a[a > 20])
```

```python
#  copy / view
a = np.array([1,2,3])
b = a.copy()
c = a.view()
print(b)    # as it
print(c)     # as it
```

```python
# Random Numbers


print(np.random.rand(3))    # random → random numbers
                            #rand → 0 se 1 ke beech ke random decimal numbers
                            # (3) → 3 numbers generate karo

print(np.random.randint(1, 100, 5))    # randint → random integers


print(np.random.randn(3,3))       # randn → Normal (Gaussian) distribution
                                  #(3,3) → 3rows, 3columns matrix

```

```python
#Linear Algebra (AI Core)


a = np.array([[1,2],
              [3,4]])

print(np.linalg.det(a))    # determination of matrix
print(np.linalg.inv(a))     # inverse of matrix
```
